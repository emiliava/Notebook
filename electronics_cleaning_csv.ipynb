```python
import pandas as pd
import numpy as np
import re
import datetime
```

```python
df_ele_1 = pd.read_excel("Amazon.co.uk _ electronics_raw_1.xlsx")
df_ele_2 = pd.read_excel("Amazon.co.uk _ electronics_raw_2.xlsx")
df_ele_3 = pd.read_excel("Amazon.co.uk _ electronics_raw_3.xlsx")
```

```python
df_ele_1.describe()
```

```python
df_ele_1.head()
```

```python
df_ele_1[['Overall Stars', "Individual Review"]]
```

```python
df_ele_1[['Overall Stars', "Individual Review"]] = df_ele_1[['Overall Stars', "Individual Review"]].apply(lambda x: x.str.rstrip(' out of 5 stars'))
df_ele_2[['Overall Stars', "Individual Review"]] = df_ele_2[['Overall Stars', "Individual Review"]].apply(lambda x: x.str.rstrip(' out of 5 stars'))
df_ele_3[['Overall Stars', "Individual Review"]] = df_ele_3[['Overall Stars', "Individual Review"]].apply(lambda x: x.str.rstrip(' out of 5 stars'))
```

```python
df_ele_1 =  df_ele_1.astype(str)
df_ele_2 =  df_ele_2.astype(str)
df_ele_3 =  df_ele_3.astype(str)
```

```python
df_ele_1['Number of Reviews'] = df_ele_1['Number of Reviews'].map(lambda x: x.rstrip(' ratings'))
df_ele_2['Number of Reviews'] = df_ele_2['Number of Reviews'].map(lambda x: x.rstrip(' ratings'))
df_ele_3['Number of Reviews'] = df_ele_3['Number of Reviews'].map(lambda x: x.rstrip(' ratings'))	
```

```python
df_ele_1.head()
#df_ele_2.head()
#df_ele_3.head()
```

```python
df_ele_1['Location and Date'] = df_ele_1['Location and Date'].apply(lambda x: x.lstrip('Reviewed in the '))
df_ele_2['Location and Date'] = df_ele_2['Location and Date'].apply(lambda x: x.lstrip('Reviewed in the '))
df_ele_3['Location and Date'] = df_ele_3['Location and Date'].apply(lambda x: x.lstrip('Reviewed in the '))
```

```python
df_ele_1.head()
```

```python
df_ele_1['Number of Reviews'] = df_ele_1['Number of Reviews'].apply(lambda x: x.replace(',',''))
df_ele_2['Number of Reviews'] = df_ele_2['Number of Reviews'].apply(lambda x: x.replace(',',''))
df_ele_3['Number of Reviews'] = df_ele_3['Number of Reviews'].apply(lambda x: x.replace(',',''))
```

```python
df_ele_1['Helpful'] = df_ele_1['Helpful'].apply(lambda x: x.replace(',',''))
df_ele_2['Helpful'] = df_ele_2['Helpful'].apply(lambda x: x.replace(',',''))
df_ele_3['Helpful'] = df_ele_3['Helpful'].apply(lambda x: x.replace(',',''))
```

```python
df_ele_2.head()
```

```python
df_ele_1["Date"] = df_ele_1["Location and Date"]
df_ele_2["Date"] = df_ele_2["Location and Date"]
df_ele_3["Date"] = df_ele_3["Location and Date"]
```

```python
df_ele_1["Date"] = df_ele_1["Date"].apply(lambda x: x.split(' on ')[-1].strip())
df_ele_2["Date"] = df_ele_2["Date"].apply(lambda x: x.split(' on ')[-1].strip())
df_ele_3["Date"] = df_ele_3["Date"].apply(lambda x: x.split(' on ')[-1].strip())
```

```python
df_ele_1.head()
```

```python
df_ele_1["Location and Date"] = df_ele_1["Location and Date"].apply(lambda x: x.split(' on ')[0].strip())
df_ele_2["Location and Date"] = df_ele_2["Location and Date"].apply(lambda x: x.split(' on ')[0].strip())
df_ele_3["Location and Date"] = df_ele_3["Location and Date"].apply(lambda x: x.split(' on ')[0].strip())
```

```python
df_ele_1.head()
```

```python
df_ele_1.rename(columns={"Location and Date": "Location"}, inplace = True)
df_ele_2.rename(columns={"Location and Date": "Location"}, inplace = True)
df_ele_3.rename(columns={"Location and Date": "Location"}, inplace = True)
```

```python
df_ele_1.head()
```

```python
df_ele_1['Price'] = df_ele_1['Price'].apply(lambda x: x.replace('\n.\n','.'))
df_ele_2['Price'] = df_ele_2['Price'].apply(lambda x: x.replace('\n.\n','.'))
df_ele_3['Price'] = df_ele_3['Price'].apply(lambda x: x.replace('\n.\n','.'))
```

```python
df_ele_1.head()
```

```python
df_ele_1['Price'] = df_ele_1['Price'].apply(lambda x: x.replace('£',''))
df_ele_2['Price'] = df_ele_2['Price'].apply(lambda x: x.replace('£',''))
df_ele_3['Price'] = df_ele_3['Price'].apply(lambda x: x.replace('£',''))
```

```python
df_ele_1.head()
```

```python
df_ele_1.info()
```

```python
df_ele_1[["day","month", "year"]] = df_ele_1["Date"].apply(lambda x: pd.Series(x.split(' ')))
df_ele_2[["day","month", "year"]] = df_ele_2["Date"].apply(lambda x: pd.Series(x.split(' ')))
df_ele_3[["day","month", "year"]] = df_ele_3["Date"].apply(lambda x: pd.Series(x.split(' ')))
```

```python
df_ele_1["Date"] = pd.to_datetime(df_ele_1['year'].astype(str)  + df_ele_1['month'] + df_ele_1["day"], format='%Y%B%d')
df_ele_2["Date"] = pd.to_datetime(df_ele_2['year'].astype(str)  + df_ele_2['month'] + df_ele_2["day"], format='%Y%B%d')
df_ele_3["Date"] = pd.to_datetime(df_ele_3['year'].astype(str)  + df_ele_3['month'] + df_ele_3["day"], format='%Y%B%d')
```

```python
df_ele_1.head()
```

```python
df_ele_1.drop(columns = ["day", "month", "year"], inplace= True)
df_ele_2.drop(columns = ["day", "month", "year"], inplace= True)
df_ele_3.drop(columns = ["day", "month", "year"], inplace= True)
```

```python
df_ele_1["Helpful"].value_counts()
```

```python
df_ele_1["Helpful"] = df_ele_1["Helpful"].apply(lambda x: x.rstrip('people found this helpful'))
df_ele_2["Helpful"] = df_ele_2["Helpful"].apply(lambda x: x.rstrip('people found this helpful'))
df_ele_3["Helpful"] = df_ele_3["Helpful"].apply(lambda x: x.rstrip('people found this helpful'))
```

```python
df_ele_1["Helpful"].replace(['One per', "na"], [1, 0], inplace = True)
df_ele_2["Helpful"].replace(['One per', "na"], [1, 0], inplace = True)
df_ele_3["Helpful"].replace(['One per', "na"], [1, 0], inplace = True)
```

```python
df_ele_1["Number of Reviews"].replace('', 0, inplace=True)
df_ele_2["Number of Reviews"].replace('', 0, inplace=True)
df_ele_3["Number of Reviews"].replace('', 0, inplace=True)
```
