```python
import pandas as pd
import numpy as np
import re
import datetime
```

```python
df_gym = pd.read_excel("../Raw(uncleaned) csv files/Amazon.co.uk _ gym equipment_raw.xlsx")
```

```python
df_gym.head()
```

```python
df_gym.info()
```

```python
df_gym[["Overall Stars", "Individual Review", "Number of Reviews", "Location and Date", "Number of Reviews", "Price"]] = df_gym[["Overall Stars", "Individual Review", "Number of Reviews", "Location and Date", "Number of Reviews", "Price"]].astype(str)
```

```python
df_gym[['Overall Stars', "Individual Review"]] = df_gym[['Overall Stars', "Individual Review"]].apply(lambda x: x.str.rstrip(' out of 5 stars'))
df_gym['Number of Reviews'] = df_gym['Number of Reviews'].map(lambda x: x.rstrip(' ratings'))
df_gym['Location and Date'] = df_gym['Location and Date'].apply(lambda x: x.lstrip('Reviewed in '))
df_gym['Number of Reviews'] = df_gym['Number of Reviews'].apply(lambda x: x.replace(',',''))

```

```python
df_gym["Name"].value_counts()
```

```python
df_gym.head()
```

```python
df_gym["Date"] = df_gym["Location and Date"]
df_gym["Date"] = df_gym["Date"].apply(lambda x: x.split(' on ')[-1].strip())
```

```python
df_gym.head()
```

```python
df_gym["Location and Date"] = df_gym["Location and Date"].apply(lambda x: x.split(' on ')[0].strip())
```

```python
df_gym.head()
```

```python
df_gym.rename(columns={"Location and Date": "Location"}, inplace = True)
```

```python
df_gym['Price'] = df_gym['Price'].apply(lambda x: x.replace('\n.\n','.'))
df_gym['Price'] = df_gym['Price'].apply(lambda x: x.replace('£',''))
```

```python
df_gym.head()
```

```python
df_gym[["day","month", "year"]] = df_gym["Date"].apply(lambda x: pd.Series(x.split(' ')))
df_gym["Date"] = pd.to_datetime(df_gym['year'].astype(str)  + df_gym['month'] + df_gym["day"], format='%Y%B%d')
df_gym.drop(columns = ["day", "month", "year"], inplace= True)
```

```python
df_gym["Location"].value_counts()
```

```python
df_gym['Location'] = df_gym['Location'].apply(lambda x: x.replace('the ',''))
```

```python
df_gym["Location"].value_counts()
```

```python
df_gym.isna().value_counts()
```

```python
df_gym.head()
```

```python
df_gym[["Product", "Title of Review", "Name", "Location", "State", "Review"]].astype(str).dtypes
df_gym[["Overall Stars", "Price", "Individual Review"]].astype(float)
df_gym[["Number of Reviews"]].astype(int)
```

```python
df_covid_dates = df_gym.loc[df_gym["Location"].isin(["United Kingdom", "United States", "Canada", "India", "Australia"])]
```

```python
df_covid_dates.describe()
```

```python
df_covid_dates = df_covid_dates[(df_covid_dates["Date"] > "2019-12-01") & (df_covid_dates['Date'] <= "2020-08-01")]
```

```python
df_covid_dates.reset_index(inplace=True, drop = True)
df_covid_dates
```

```python
df_covid_dates = df_covid_dates.drop(columns=["Overall Stars", "Number of Reviews", "Title of Review", "Product Description", "State", "Name"])
df_covid_dates
```
